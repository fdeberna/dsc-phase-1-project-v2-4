```python
### IMDB seems a very complete databse, with info about actors in each movie, etc, so let's start by fetching the basic info
q = "SELECT * FROM movie_basics"
cur.execute(q)
```

```python
basic = cur.fetchall()
```

```python
df_movie_basic = pd.DataFrame(basic,columns = [column_id[0] for column_id in cur.description])
```

```python
df_movie_basic
```

```python
### going to lower the titles to avoud issues, for the budgets df also
df_movie_basic['title_lower'] = [x.lower() for x in df_movie_basic.primary_title]
df_budgets['movie_lower'] = [x.lower() for x in df_budgets.movie]
```

```python
df_budgets
```

```python
df_movie_basic.merge(df_budgets, left_on = 'title_lower', right_on = 'movie_lower')
```

```python
### let's look at actors
q = """SELECT * FROM movie_basics
     JOIN principals ON movie_basics.movie_id = principals.movie_id
     JOIN persons    ON principals.person_id = persons.person_id;"""
cur.execute(q)
df_movie_actors = pd.DataFrame(cur.fetchall())
```
